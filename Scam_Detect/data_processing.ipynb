```python
import pandas as pd
import numpy as np
import random
```

# Thai dataset

```python
df = pd.read_csv('original_data/Original_Dataset.csv')
df.head(10)
```

```python
# edit columns
df_new = df.drop('Id', axis=1)
df_new = df_new.rename(columns={'Text': 'text', 'Type': 'label'})
df_new.head(10)
```

```python
# make label to be either 1 (Scam) or 0 (Other)
df_new['label'] = np.where(df_new['label'] == 'Scam', 1, 0)
df_new.head(10)
```

```python
# save the formatted only version to csv

df_new.to_csv('processed_data/formatted_only.csv', index=False)
```

## Text/Data Processing

```python
# Removing duplicates conversation

print(len(df_new))
df_new.drop_duplicates(subset=['text'], inplace=True)
print(len(df_new))
```

### assuming audio is from both side of the conversation

```python
df_both = df_new.copy()
df_both['text'] = df_both['text'].str.strip()
df_both['text'] = df_both['text'].str.replace("[Call]", "")
df_both['text'] = df_both['text'].str.replace("[User]", "")
df_both['text'] = df_both['text'].str.replace("(", "")
df_both['text'] = df_both['text'].str.replace(")", "")
df_both['text'].head(20)
```

```python
df_both.to_csv('processed_data/both.csv', index=False)
```

### assuming audio is only from the other side of the conversation

```python
df_caller = df_new.copy()

pattern = r"\[Call\]\s*\((.*?)\)"
df_caller['text'] = df_caller['text'].str.findall(pattern) #.str.join('')

df_caller_exploded = df_caller.explode('text')

df_caller_joined = df_caller.copy()
df_caller_joined['text'] = df_caller['text'].str.join(' ')

df_caller_exploded.to_csv('processed_data/caller_exploded.csv', index=False)
df_caller_joined.to_csv('processed_data/caller_joined.csv', index=False)

print(df_caller_exploded.head(5))
print()
print(df_caller_joined.head(5))
```

# From English/AI parsed dataset

```python
with open('original_data/archive/Thai_NonScam_Gemini.txt', 'r', encoding='utf-8') as f:
    nonscam_lines = [line.strip() for line in f.readlines()]

with open('original_data/archive/Thai_Scam_Gemini.txt', 'r', encoding='utf-8') as f:
    scam_lines = [line.strip() for line in f.readlines()]

label = [0 for x in nonscam_lines] + [1 for x in scam_lines]

data = {
    "text": nonscam_lines + scam_lines,
    "label": label
}

df_ai = pd.DataFrame(data)
df_ai = df_ai.sample(frac=1).reset_index(drop=True)

greetings = ["สวัสดีครับ","สวัสดีค่ะ",""]

def generate_bank_no():
    length = random.randint(10,14)
    digits = ''.join(random.choices("0123456789", k=length))

    indices = sorted(random.sample(range(1, length), random.randint(0,3)))

    parts = []
    last_idx = 0
    for idx in indices:
        parts.append(digits[last_idx:idx])
        last_idx = idx
    parts.append(digits[last_idx:])
    
    return "-".join(parts)

names = [
    "สมชาย", "สมศักดิ์", "อนุรักษ์", "กิตติชัย", "ไชยา", 
    "อาทิตย์", "ณัฐพงศ์", "ศราวุธ", "สันติ", "เกษม",
    "มะลิ", "กัญญา", "อารี", "รัตนา", "อัญชลี", 
    "ลออ", "ศิริพร", "บุษบา", "สุนารี", "จันทนา"
]

name_pattern = r"\[ชื่อ.*?\]"

df_ai['text'] = df_ai['text'].str.replace(
    name_pattern, 
    lambda m: random.choice(names), 
    regex=True
)

money_pattern = r"\[จำนวนเงิน.*?\]"

df_ai['text'] = df_ai['text'].str.replace(
    money_pattern, 
    lambda m: f"{round(random.randint(1000,30000), -2)}", 
    regex=True
)

day_pattern = r"\[วันที่\]"

df_ai['text'] = df_ai['text'].str.replace(
    day_pattern, 
    lambda m: f"{random.randint(1,31)}", 
    regex=True
)

df_ai['text'] = df_ai['text'].apply(lambda x: x.replace("[สวัสดีครับ/ค่ะ]", random.choice(greetings)))
df_ai['text'] = df_ai['text'].apply(lambda x: x.replace("[หมายเลขบัญชี]", generate_bank_no()))
df_ai['text'] = df_ai['text'].str.replace(r'[()]', '', regex=True)
df_ai['text'] = df_ai['text'].str.replace("...", "")
df_ai['text'] = df_ai['text'].str.replace('"', "")
df_ai['text'] = df_ai['text'].str.replace('"', "")

else_pattern = r"\[.*\]"

df_ai['text'] = df_ai['text'].str.replace(
    else_pattern, 
    lambda m: "", 
    regex=True
)

df_ai.to_csv('processed_data/ai.csv', index=False)
print("\n".join(df_ai[df_ai['text'].str.contains("[", regex=False)]['text'].tolist()))
# print("\n".join(df_ai[df_ai['text'].str.contains("\[*\]")]['text'].tolist()))
```

```python
df_ai_and_caller_exploded = pd.concat([df_caller_exploded, df_ai], axis=0, ignore_index=True).sample(frac=1).reset_index(drop=True)
df_ai_and_caller_exploded.to_csv('processed_data/ai_and_caller_exploded.csv', index=False)
```
