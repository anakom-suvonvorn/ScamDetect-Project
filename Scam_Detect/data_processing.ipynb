# Formatting

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Original_Dataset.csv')
df.head(10)
```

```python
# edit columns
df_new = df.drop('Id', axis=1)
df_new = df_new.rename(columns={'Text': 'text', 'Type': 'label'})
df_new.head(10)
```

```python
# make label to be either 1 (Scam) or 0 (Other)
df_new['label'] = np.where(df_new['label'] == 'Scam', 1, 0)
df_new.head(10)
```

```python
# save the formatted only version to csv

df_new.to_csv('processed_data/formatted_only.csv', index=False)
```

# Text/Data Processing

```python
# Removing duplicates conversation

print(len(df_new))
df_new.drop_duplicates(subset=['text'], inplace=True)
print(len(df_new))
```

## assuming audio is from both side of the conversation

```python
df_both = df_new.copy()
df_both['text'] = df_both['text'].str.strip()
df_both['text'] = df_both['text'].str.replace("[Call]", "")
df_both['text'] = df_both['text'].str.replace("[User]", "")
df_both['text'] = df_both['text'].str.replace("(", "")
df_both['text'] = df_both['text'].str.replace(")", "")
df_both['text'].head(20)
```

```python
df_both.to_csv('processed_data/both.csv', index=False)
```

## assuming audio is only from the other side of the conversation

```python
df_caller = df_new.copy()

pattern = r"\[Call\]\s*\((.*?)\)"
df_caller['text'] = df_caller['text'].str.findall(pattern) #.str.join('')

df_caller_exploded = df_caller.explode('text')

df_caller_joined = df_caller.copy()
df_caller_joined['text'] = df_caller['text'].str.join(' ')

df_caller_exploded.to_csv('processed_data/caller_exploded.csv', index=False)
df_caller_joined.to_csv('processed_data/caller_joined.csv', index=False)

print(df_caller_exploded.head(5))
print()
print(df_caller_joined.head(5))
```
